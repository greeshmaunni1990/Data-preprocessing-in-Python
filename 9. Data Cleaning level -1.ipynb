```python
from os import listdir
import pandas as pd
import io
```

#### Unwise data collection

```python
# Step 1
# List the data
filenames = listdir('Speeches')
print(filenames)
```

```python
# Step 2
# Create a DataFrame that hold the value
speech_df = pd.DataFrame(index=range(len(filenames)), columns=['File Name', 'Content'])
print(speech_df)
```

```python
for i, f_name in enumerate(filenames):
    f = io.open('Speeches/'+f_name, "r", encoding='utf-8')
    f_content = f.readlines()
    speech_df.at[i, 'File Name'] = f_name
    speech_df.at[i, 'Content'] = f_content[0]
    
speech_df
```

```python
print(speech_df['File Name'])
print(speech_df['Content'])
```

```python
print(speech_df.File Name) # This error is because of the space in File Name
```

```python
# Change the column name
speech_df.columns = ['FileName', 'Content']
```

```python
print(speech_df.FileName)
```
