```python
import pandas as pd
from os import listdir
import io
```

```python
filenames = listdir('Speeches')
print(filenames)

speech_df = pd.DataFrame(index=range(len(filenames)), columns=['File Name', 'Content'])


for i, f_name in enumerate(filenames):
    f = io.open('Speeches/'+f_name, "r", encoding='utf-8')
    f_content = f.readlines()
    speech_df.at[i, 'File Name'] = f_name
    speech_df.at[i, 'Content'] = f_content[0]
speech_df.columns = ['FileName', 'Content']
print(speech_df)
```

#### Unpack FileName

```python
Months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Oct','Sep','Nov','Dec']
def SeperateCity(v):
    for mon in Months:
        if (mon in v):
            return v[:v.find(mon)]

speech_df['City'] = speech_df.FileName.apply(SeperateCity)
print(speech_df)
```

```python
def separateDate(r):
    return r.FileName[len(r.City):r.FileName.find('.txt')]

speech_df['Date'] = speech_df.apply(separateDate, axis=1)
speech_df.head()
```

```python
# convert DATE column from string to Date object
speech_df.Date = pd.to_datetime(speech_df.Date, format='%b%d_%Y')
speech_df
```

```python
def extract_year_month_day(r):
    r['Year'] = r.Date.year
    r['Month'] = r.Date.month
    r['Day'] = r.Date.day
    return r
speech_df = speech_df.apply(extract_year_month_day, axis=1)
```

```python
# Drop FileName column as all the information is created as separate columns
speech_df.drop(['FileName'], axis =1, inplace=True)
```

#### Unpack Content

```python
Words = ['vote','tax','campaign','economy']

def FindWordRatio(row):
    total_n_words = len(row.Content.split(' '))
    for w in Words:
        row['r_{}'.format(w)] = row.Content.count(w)/total_n_words
    return row

speech_df = speech_df.apply(FindWordRatio,axis=1)
```
