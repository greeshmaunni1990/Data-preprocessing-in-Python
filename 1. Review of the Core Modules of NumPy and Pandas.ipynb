### Applying function to Series

```python
def multiply_by_two(n):
    return n * 2

adult_df['age'].apply(multiply_by_two)
```

#### Applying a function  Analytic example 1

```python
# 1. First, total_fnlwgt, which is the sum of all the values in the fnlwgt column, is calculated.
total_fnlwgt = adult_df['fnlwgt'].sum()
print(total_fnlwgt)

# 2. Second, the CalculatePercentage function is defined. This function outputs the input values divided by total_fnlwgt
# and multiplied by 100 (to develop a percentage).

def CalculatePercentage(n):
    return n / total_fnlwgt * 100

# 3. Third, the CalculatePercentage function is applied to the series adult_df.fnlwgt.

adult_df['fnlwgt'].apply(CalculatePercentage)
adult_df.head()
```

#### Applying lambda function

```python
total_fnlwgt = adult_df['fnlwgt'].sum()
adult_df['fnlwgt'] = adult_df['fnlwgt'].apply(lambda n: n/total_fnlwgt * 100)
adult_df.head()
```

### Applying function to DataFrame

```python
# 1. First, when defining the CalcLifeNoEd() function, the input row was assumed to be a row series of adult_df. 
# In other words, the CalcLifeNoEd() function is tailored just for application to adult_df or 
# any DataFrame that has age and eduction_num as columns.

def CalcLifeNoEd(row):
    return row.age - row.education_num

# axis = 1 will be applied to every rows 
# axis = 0 will be applied to every columns
adult_df.apply(CalcLifeNoEd, axis=1).head(10)
```

### Applying a function  Analytic example 2

#### Which one is more important in terms of your financial success: education or life experience?

```python
# Create 2 columns
#    * LifeNoEd -> The number of years for which you have lived without formal education
#    * capitalNet -> Difference between CapitalGain and CapitalLoss

# Then find out if education_num or LifeNoEd has higher correlation to capitalNet

adult_df['LifeNoEd'] = adult_df.apply(lambda row: row.age - row.education_num, axis=1)
adult_df['capitalNet'] = adult_df.apply(lambda row: row.capitalgain - row.capitalloss, axis=1)
```

```python
adult_df.head(5)
```
