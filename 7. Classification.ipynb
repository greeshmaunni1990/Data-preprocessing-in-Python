```python
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import numpy as np
```

```python
applicant_df = pd.read_csv('CustomerLoan.csv')
applicant_df
```

#### K-Nearest Neighbour

```python
# Normalize data
applicant_df['Normalized Income'] = (applicant_df['income'] - applicant_df['income'].min())/(applicant_df['income'].max() - applicant_df['income'].min())

applicant_df['Normalized Score'] = (applicant_df['score'] - applicant_df['score'].min())/(applicant_df['score'].max() - applicant_df['score'].min())

applicant_df
```

```python
predictors = ['Normalized Income', 'Normalized Score']
target = 'default'

data_x = applicant_df[predictors].drop(index=[20])
data_y = applicant_df[target].drop(index=[20])
```

```python
knn =KNeighborsClassifier(n_neighbors=4)
knn.fit(data_x, data_y)
```

```python
new_data = pd.DataFrame({'Normalized Income': applicant_df['Normalized Income'].iloc[20],
                        'Normalized Score': applicant_df['Normalized Score'].iloc[20]}, index=[20])

new_data
```

```python
knn.predict(new_data)
```

#### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
predictors = ['income', 'score']
target = 'default'

data_x = applicant_df[predictors].drop(index=[20])
data_y = applicant_df[target].drop(index=[20])
```

```python
decision_tree_class = DecisionTreeClassifier()
decision_tree_class.fit(data_x, data_y)
```

```python
decision_tree_class.predict(new_data)
```
