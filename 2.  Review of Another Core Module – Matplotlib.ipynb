```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
adult_df = pd.read_csv('adult.csv')
```

#### Histogram

```python
# used for numerical values
plt.hist(adult_df.age)
```

#### Boxplot

```python
# used for numerical values
plt.boxplot(adult_df.age, vert=False)
plt.show()
```

```python
amazon_df = pd.read_csv('Amazon Stock.csv')
apple_df = pd.read_csv('Apple Stock.csv')

print(amazon_df.head())
print(apple_df.head())
```

#### Line plot

```python
# used for time series data where the trend is changing.
plt.plot(amazon_df.Close)
plt.plot(apple_df.Close)
plt.show()
```

#### Scatter plot

```python
# This function is great for examining the relationship between numerical attributes
plt.scatter(amazon_df.Close, apple_df.Close)
plt.show()
```

### Modifying the visuals

#### Add title and ylabel

```python
plt.plot(amazon_df.Close)
plt.plot(apple_df.Close)
plt.title('Line plot of Amazon and Apple stock from 2000 to 2020')
plt.ylabel('Closing Price')
plt.show()
```

#### Add legend

```python
plt.plot(amazon_df.Close, label='Amazon')
plt.plot(apple_df.Close, label='Apple')
plt.title('Line plot of Amazon and Apple stock from 2000 to 2020')
plt.ylabel('Closing Price')
plt.legend()
plt.show()
```

#### Adding ticks

```python
plt.plot(amazon_df.Close, label='Amazon')
plt.plot(apple_df.Close, label='Apple')
plt.title('Line plot of Amazon and Apple stock from 2000 to 2020')
plt.ylabel('Closing Price')
plt.xticks([0, 500, 1000, 1500, 2000, 2500, 3000, 3500, 4000, 4500, 5000, 5500], rotation=90)
plt.legend()
plt.show()
```
