#### Summarize Numerical Attribute

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
adult_df = pd.read_csv('adult.csv')
numerical_attributes = ['age', 'fnlwgt', 'education-num', 'capitalGain', 'capitalLoss',
'hoursPerWeek']

for attr in numerical_attributes:
    plt.subplot(2, 1, 1)
    adult_df[attr].plot.hist()
    plt.subplot(2, 1, 2)
    adult_df[attr].plot.box(vert=False)
    plt.tight_layout()
    plt.show()
#     plt.savefig('{}.png'.format(attr), dpi=600)
```

#### Summarize Categorical Attribute

```python
categorical_attribute = ['workclass', 'education', 'marital-status', 'occupation',
'relationship', 'race', 'sex','nativeCountry','income']

for attr in categorical_attribute:
    adult_df[attr].value_counts().plot.barh()
    plt.title(attr)
    plt.tight_layout()
    plt.show()
```
